```python
import gradio as gr
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Configuration parameters
MODEL_NAME = "gpt2"  # Consider upgrading to a more advanced model
MAX_LENGTH = 100
TEMPERATURE = 0.7
TOP_P = 0.9
REPETITION_PENALTY = 1.2
INAPPROPRIATE_WORDS = ["shit", "crap", "damn"]

# Load the pre-trained GPT-2 model and tokenizer
model = GPT2LMHeadModel.from_pretrained(MODEL_NAME)
tokenizer = GPT2Tokenizer.from_pretrained(MODEL_NAME)

def generate_response(input_text):
    """
    Generate a response using the GPT-2 model based on the input text.

    Parameters:
    - input_text (str): The input text to generate a response for.

    Returns:
    - str: The generated response or an error message if an exception occurs.
    """
    try:
        # Encode the input text
        input_ids = tokenizer.encode(input_text, return_tensors='pt')

        # Generate text using the GPT-2 model
        output = model.generate(
            input_ids,
            max_length=MAX_LENGTH,
            num_return_sequences=1,
            temperature=TEMPERATURE,
            top_p=TOP_P,
            repetition_penalty=REPETITION_PENALTY
        )

        # Decode the generated text
        generated_text = tokenizer.decode(output[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)

        # Filter out inappropriate content
        if any(word in generated_text.lower() for word in INAPPROPRIATE_WORDS):
            return "The generated text contains inappropriate content. Please try again with a different input."

        # Improve relevance: Basic check for accuracy in factual answers
        if "continent" in input_text.lower() and ("number" in input_text.lower() or "how many" in input_text.lower()):
            if "seven" not in generated_text.lower():
                return "The correct answer is that there are seven continents in the world."

        return generated_text

    except Exception as e:
        return f"An error occurred: {str(e)}"

# Create the Gradio interface
with gr.Blocks() as demo:
    input_textbox = gr.Textbox(lines=3, placeholder="Type your text here...")
    output_textbox = gr.Textbox(label="Response", lines=5, interactive=False)
    submit_button = gr.Button("Submit")

    def handle_submit(input_text):
        """
        Handle the submission of the input text by generating a response.

        Parameters:
        - input_text (str): The input text submitted by the user.

        Returns:
        - str: The generated response.
        """
        response = generate_response(input_text)
        return response

    submit_button.click(handle_submit, inputs=input_textbox, outputs=output_textbox)

# Launch the interface
if __name__ == "__main__":
    demo.launch()
```
